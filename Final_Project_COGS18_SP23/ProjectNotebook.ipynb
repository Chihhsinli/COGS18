# Project Description

The original intention of this program design is to help children practice arithmetic. Generating arithmetic questions for children and checking their answers can be tedious for adults. Therefore, I have created this program to assist children in their arithmetic practice.

The program is divided into addition and subtraction mode, as well as multiplication and division mode. It offers both an easy mode and a difficult mode. The easy mode involves arithmetic operations with two numbers within 10, while the difficult mode extends to operations with two numbers within 100.

To make arithmetic less suffering for children, all division operations result in whole numbers. Once you have made your selection, you can start practicing. Each practice session consists of 10 questions, and upon completion, you will receive the number of incorrect answers along with the correct solutions.

```python
# Do a bunch of things.
def fun():
    while True:
        print("Select mode:")
        print("1. Addition/Subtraction")
        print("2. Multiplication/Division")
        print("3. Quit")
        choice = input("Enter your choice (only the number):")

        if choice == "1":
            mode = "Addition/Subtraction"
        elif choice == "2":
            mode = "Multiplication/Division"
        elif choice == "3":
            print("Thank you for using. Have fun!")
            break
        else:
            print("Invalid choice. Please try again!")
            continue

        print("Select difficulty:")
        print("1. Easy")
        print("2. Hard")
        difficulty_choice = input("Enter your choice:")

        if difficulty_choice == "1":
            difficulty = "Easy"
        elif difficulty_choice == "2":
            difficulty = "Hard"
        else:
            print("Invalid choice. Please try again!")
            continue

        num_questions = 10
        wrong_answers = []

        for _ in range(num_questions):
            question, answer = generate_question(mode, difficulty)
            user_answer = input(question)

            try:
                user_answer = int(user_answer)
            except ValueError:
                pass

            if user_answer != answer:
                wrong_answers.append((question, answer))

        print("\n===== Results =====")
        print(f"Number of wrong answers: {len(wrong_answers)}\n")

        if len(wrong_answers) > 0:
            print("Wrong answers:")
            for question, answer in wrong_answers:
                print(f"{question}Correct answer: {answer}\n")
        
        else:
            print("You got all the questions right. Congratulations!\n")
```

```python
# test it out
fun()
```
