```
import random

def generate_addition_subtraction_question(difficulty):
    if difficulty == "Easy":
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)
    elif difficulty == "Hard":
        num1 = random.randint(1, 100)
        num2 = random.randint(1, 100)
    else:
        raise ValueError("Invalid difficulty level")

    operator = random.choice(['+', '-'])
    if operator == '+':
        answer = num1 + num2
    else:
        num1, num2 = max(num1, num2), min(num1, num2)
        answer = num1 - num2
    question = f"What is {num1} {operator} {num2}? "
    return question, answer

def generate_multiplication_division_question(difficulty):
    while True:
        if difficulty == "Easy":
            num1 = random.randint(1, 10)
            num2 = random.randint(1, 10)
        elif difficulty == "Hard":
            num1 = random.randint(1, 100)
            num2 = random.randint(1, 100)
        else:
            raise ValueError("Invalid difficulty level")

        operator = random.choice(['*', '/'])
        if operator == '*':
            answer = num1 * num2
        else:
            num1, num2 = num1 * num2, num1
            answer = num1 // num2
        
        if num1 <= 1000 and num2 <= 1000:
            break

    question = f"What is {num1} {operator} {num2}? "
    return question, answer
```

```
def generate_question(mode, difficulty):
    if mode == "Addition/Subtraction":
        return generate_addition_subtraction_question(difficulty)
    elif mode == "Multiplication/Division":
        return generate_multiplication_division_question(difficulty)
    else:
        raise ValueError("Invalid mode")
```

```
def fun():
    while True:
        print("Select mode:")
        print("1. Addition/Subtraction")
        print("2. Multiplication/Division")
        print("3. Quit")
        choice = input("Enter your choice (only the number):")

        if choice == "1":
            mode = "Addition/Subtraction"
        elif choice == "2":
            mode = "Multiplication/Division"
        elif choice == "3":
            print("Thank you for using. Have fun!")
            break
        else:
            print("Invalid choice. Please try again!")
            continue

        print("Select difficulty:")
        print("1. Easy")
        print("2. Hard")
        difficulty_choice = input("Enter your choice:")

        if difficulty_choice == "1":
            difficulty = "Easy"
        elif difficulty_choice == "2":
            difficulty = "Hard"
        else:
            print("Invalid choice. Please try again!")
            continue

        num_questions = 10
        wrong_answers = []

        for _ in range(num_questions):
            question, answer = generate_question(mode, difficulty)
            user_answer = input(question)

            try:
                user_answer = int(user_answer)
            except ValueError:
                pass

            if user_answer != answer:
                wrong_answers.append((question, answer))

        print("\n===== Results =====")
        print(f"Number of wrong answers: {len(wrong_answers)}\n")

        if len(wrong_answers) > 0:
            print("Wrong answers:")
            for question, answer in wrong_answers:
                print(f"{question}Correct answer: {answer}")
        
        else:
            print("You got all the questions right. Congratulations!\n")
```
